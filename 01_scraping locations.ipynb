```python
from bs4 import BeautifulSoup
import requests
from csv import writer
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
import re
```

```python
url = 'https://www.wogibtswas.at/fl/72801-denns-biomarkt-filialen?page=1'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
tags = soup.find_all("address", class_="address-store")
    
with open ('denns1.txt', 'w', encoding = 'utf8') as f:
    for tag in tags:
        split_strings = tag.text.split('\n')
        non_empty_strings = [s.strip() for s in split_strings if s.strip()]
        reversed_strings = reversed(non_empty_strings)
        result_string = ', '.join(reversed_strings)
        f.write(result_string)
        f.write("\n")
        
url = 'https://www.wogibtswas.at/fl/72801-denns-biomarkt-filialen?page=2'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
tags = soup.find_all("address", class_="address-store")
    
with open ('denns2.txt', 'w', encoding = 'utf8') as f:
    for tag in tags:
        split_strings = tag.text.split('\n')
        non_empty_strings = [s.strip() for s in split_strings if s.strip()]
        reversed_strings = reversed(non_empty_strings)
        result_string = ', '.join(reversed_strings)
        f.write(result_string)
        f.write("\n")
        
url = 'https://www.wogibtswas.at/fl/72801-denns-biomarkt-filialen?page=3'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
tags = soup.find_all("address", class_="address-store")
    
with open ('denns3.txt', 'w', encoding = 'utf8') as f:
    for tag in tags:
        split_strings = tag.text.split('\n')
        non_empty_strings = [s.strip() for s in split_strings if s.strip()]
        reversed_strings = reversed(non_empty_strings)
        result_string = ', '.join(reversed_strings)
        f.write(result_string)
        f.write("\n")
```

```python
# Unibox Adrerssen
driver = webdriver.Chrome()

driver.get("https://unimarkt.at/standorte-alt/alle-standorte/")
driver.find_element(By.XPATH, '//*[@id="BorlabsCookieBox"]/div/div/div/div[1]/div/div/div/p[2]/a').click()
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="gb-form"]/div/button[1]').click()

links = driver.find_elements(By.PARTIAL_LINK_TEXT, 'UNIBOX')
addresses_uniboxnew = []

for link in links:
    link.click()
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    address = soup.find("address").text
    addresses_uniboxnew.append(address.split("\n")[1])
    driver.back()
    time.sleep(2)

driver.quit()
```

```python
# Unimarkt Adressen
driver = webdriver.Chrome()

driver.get("https://unimarkt.at/standorte-alt/alle-standorte/")
driver.find_element(By.XPATH, '//*[@id="BorlabsCookieBox"]/div/div/div/div[1]/div/div/div/p[2]/a').click()
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="gb-form"]/div/button[1]').click()

links = driver.find_elements(By.PARTIAL_LINK_TEXT, 'UNIMARKT')
addresses_unimarktnew = []

for link in links:
    link.click()
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    address = soup.find("address").text
    addresses_unimarktnew.append(address.split("\n")[1])
    driver.back()
    time.sleep(2)
    
    
driver.quit()
```

```python
# Write Unibox addresses in correct format
with open ('unibox.txt', 'w', encoding = 'utf8') as f:
    for box in addresses_uniboxnew:
        input_string = box
        match = re.search(r'\d{4}', input_string)
        index = match.start()
        part1 = input_string[:index]
        part2 = input_string[index:]
        res = part2 + ", " + part1
        result = res[:(len(res)-1)]
        f.write(result)
        f.write("\n")
```

```python
driver = webdriver.Chrome()

driver.get("https://filialen.hofer.at/")
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="onetrust-accept-btn-handler"]').click()
links = ['https://filialen.hofer.at/burgenland',
        'https://filialen.hofer.at/kaernten',
        'https://filialen.hofer.at/niederoesterreich',
        'https://filialen.hofer.at/oberoesterreich',
        'https://filialen.hofer.at/salzburg',
        'https://filialen.hofer.at/steiermark',
        'https://filialen.hofer.at/tirol',
        'https://filialen.hofer.at/vorarlberg'
        ]
addresses = []

# get all links

for link in links:
    time.sleep(5)
    driver.get(link)
    time.sleep(5)
    bezirklinks = driver.find_elements(By.CLASS_NAME, 'Directory-listLink')
    
    for i in range(len(bezirklinks)):
        bezirklinks = driver.find_elements(By.CLASS_NAME, 'Directory-listLinkText')
        bezirklinks[i].click()
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        try:
            driver.find_element(By.CLASS_NAME, 'Core-paymentOption')
            address = soup.find_all("address", id="address")
        
        except NoSuchElementException:
            address = soup.find_all("address")
            
        for ad in address:
            addresses.append(ad.text)
            
        driver.back()
        time.sleep(1)

driver.quit()        
```
